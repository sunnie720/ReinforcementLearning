```python
import numpy as np
import pandas as pd 
import random
import warnings 
warnings.filterwarnings(action='ignore')
```

```python
# 환경 세팅
actions = ["삼전","네이비","삼풍"] # 행동들
mean = {"삼전":4000, "네이비":3000, "삼풍":-2000}  # 보상의 모평균(실제 가치)
sd = {"삼전":5000, "네이비":10000, "삼풍":70000}   # 보상의 모표준편차

first_rewards = pd.Series([-4000,1000,100000], index=actions)  # 보상 = R_1
first_counts = pd.Series([0,0,1], index=actions)               # 횟수 = 1
first_values = pd.Series([-4000, 1000, 100000], index=actions) # 가치 = Q_1 (a)
first = pd.DataFrame([first_rewards, first_counts, first_values], index=['first_rewards','first_counts', 'first_values']).T
first
```

```python
# 그리디 
def greedy(n_iter, actions, first, seed):
    random.seed(seed)
    R = first['first_rewards']
    N = first['first_counts']
    Q = first['first_values']
    # 탐욕 알고리즘-------------------------------------------
    for t in range(1, 2*n_iter+1):
        best = Q.idxmax() #행동 = A_t (ex. '삼풍')
        R_t = random.normalvariate(mean[best], sd[best]) #R_t
        R[best] = R[best] + R_t    # 보상 누적
        N[best] = N[best] + 1      # 횟수 누적
        Q[best] = R[best] / N[best] # 가치 계산 (산술평균)
        if t%10==0:
            print(t,"번째")
            print(N)
        t += 1
    return pd.DataFrame([N, R, Q], index=['counts(N)','rewards(R)', 'values(Q)']).T

greedy(100, actions, first, seed=5)
```

```python
# 그리디 방법 보상 총합
print(int(-4000.000000+645149.342989-82687.253425), "원")
```

```python
# Epsilon-Greedy
def eGreedy(n_iter, actions, first, eps, seed):
    random.seed(seed)
    R = first['first_rewards']
    N = first['first_counts']
    Q = first['first_values']
    for t in range(1, 2*n_iter+1):
        best = Q.idxmax() if eps < random.random() else random.choice(actions)
        # 보상 업데이트
        R_t = random.normalvariate(mean[best], sd[best]) #R_t
        R[best] = R[best] + R_t
        N[best] = N[best] + 1 # 횟수
        Q[best] = R[best] / N[best] # 가치 계산 (산술평균)
        if t%10==0:
            print(t,"번째")
            print(N)
        t += 1
    return pd.DataFrame([N, R, Q], index=['counts(N)','rewards(R)', 'values(Q)']).T

eGreedy(100, actions, first, eps=0.3, seed=5)
```

```python
# Upper Confidence Bound
def UCB(n_iter, first, control, seed):
    random.seed(seed)
    R = first['first_rewards']
    N = first['first_counts']
    Q = first['first_values']
    for t in range(1, 2*n_iter+1):
        # UCB 알고리즘으로 best 행동을 선택
        best = Q + control * np.sqrt(np.log(t) / N)
        best = best.idxmax()
        # 보상 업데이트
        R_t = random.normalvariate(mean[best], sd[best]) #R_t
        R[best] = R[best] + R_t
        N[best] = N[best] + 1 # 횟수
        Q[best] = R[best] / N[best] # 가치 계산 (산술평균)
        if t%10==0:
            print(t,"번째")
            print(N)
        t += 1
    return pd.DataFrame([N, R, Q], index=['counts(N)','rewards(R)', 'values(Q)']).T

UCB(100, first, control=2, seed=5)
```

```python
# Thompson Sampling - R코드
Thompson = function(n_iter,actions,first,probs, seed){
  
  set.seed(seed)
  n_actions = length(actions)
  rewards = ifelse (first$first_rewards>0, 1,0) # 보상을 받으면 1, 못받으면 0
  counts = first$first_counts
  values = first$first_values

  for (t in 2:n_iter) {
    B<-matrix(rbeta(n_actions, rewards+1, (counts-rewards)+1),1, byrow = TRUE)
    P<-table(factor(max.col(B), levels=1:ncol(B)))/nrow(B)
    # tmp are the weights for each time step
    tmp<-round(P*1,0)
    
    # Update the Rewards
    rewards = rewards+rbinom(rep(1,n_actions), size=P, prob = probs)
    
    #Update counts
    counts = counts+tmp
    
  }
  return(setNames(counts, actions))
}

# 삼전, 네이비, 삼풍 주식을 통해 보상을 얻을 확률 <- 모집단 정규분포에서 각 1,000번 난수를 뽑아서 0보다 클 확률로 설정
set.seed(5)
probs<-c(sum(rnorm(1000, mean['삼전'], sd['삼전'])>0) / 1000,
         sum(rnorm(1000, mean['네이비'], sd['네이비'])>0) / 1000,
         sum(rnorm(1000, mean['삼풍'], sd['삼풍'])>0) / 1000)

probs= setNames(probs,actions)
probs
Thompson(100, actions,first, probs, seed=5)
```
